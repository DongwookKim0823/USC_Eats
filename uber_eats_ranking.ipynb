```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('C:\\Users\\82107\\Downloads\\restaurants.csv')
```

```python
df = df.dropna(subset = ['full_address', 'price_range'])    # 가격범위 또는 주소에 결측값이 포함된 행 제거
```

```python
df['state'] = df.full_address.str.split(',').str[-2].str.strip()    # 주소값에서 주(state) 정보 추출 후 컬럼으로 추가
```

```python
df['price_num_range'] = df.price_range.str.len()    # $문자를 숫자(1~4로 변환 후 컬럼에 추가)
```

```python
aaa = df[df['state'].str.contains('DC')].index    # 워싱턴 DC행 인덱스 추출
df = df.drop(aaa)    #워싱턴 DC 행 제거
```

```python
df.groupby(df['state']).describe()    #(워싱턴DC를 제외한)주별로 묶고 price num range($의 개수)의 평균 구하기
```

```python
stopwords = ['family','american','food','and','&amp;','+','/','whatever','to','fast','friendly','asian','healthy','mexican','meal','breakfast','brunch','comfort','convenience','everyday','essential','chinese','personal','care','home']
state_name = df.groupby(df['state']).describe().index.values    # 주별 인덱스 추출
result = {}

for i in state_name:

    df_ = df[df['state'] == i]
    df_to_list = df_['category'].str.split(',').dropna().to_list()    # ','로 분할 후 결측값 제거하여 리스트로 변환

    category = []    # 소문자로 변환, 공백 제거, 공백으로 분할, 끝에 's' 제거 후 리스트에 추가
    for h in df_to_list:
        for j in h:
            temp = j.lower().strip().split()

            for k in temp:
                if k[-1] == 's':
                    category.append(k[:-1])
                else:
                    category.append(k)
            
    category_counts = {}
    for x in category:    # 딕셔너리로 메뉴 개수 세기
        if x in stopwords:
            continue

        if x in category_counts:
            category_counts[x] += 1
        else:
            category_counts[x] = 1

    if 'sandwiche' in category_counts:    # sandwiche와 sandwich 벨류값 합치기
        category_counts['sandwich'] += category_counts['sandwiche']
        category_counts.pop('sandwiche') 


    dict_category = sorted(category_counts.items(), key=lambda x: x[1], reverse=True)    # 벨류값 내림차순으로 정렬
    
    count = 0
    temp_list = []
    for m in dict_category:
        if(count == 3):
            break
        temp_list.append(m[0])
        count += 1
    
    result[i] = temp_list
        
```

```python
print(result)
```

```python
output = pd.DataFrame.from_dict(result)
output
```

```python
output.to_csv("menu_ranking_list.csv", mode='a', header=True)
```
